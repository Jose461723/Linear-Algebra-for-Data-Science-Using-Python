```python
#Importing libraries
import pandas as pd #converts dataset to pandas dataframe 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
#Converting dataset into a pandas dataframe 
titanic_df = pd.read_csv('Titanic-Dataset.csv')
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x='Survived', data=titanic_df,hue='Pclass', palette='winter')
```

```python
#Histogram 
plt.figure(figsize=(8,6))
sns.histplot(titanic_df['Age'])
```

```python
#Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='Pclass' ,y='Age', data=titanic_df, palette='Accent', linewidth=3)
```
