Using graphs in Python to model Data

```python
#Importing libraries
import numpy as np
import random
import sympy as sy
import pandas as pd #converts dataset to pandas dataframe 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Converting dataset into a pandas dataframe
```

```python
DataScience_df = pd.read_csv('DataScience_salaries_2024.csv')
```

```python
DataScience_df.head()
```

```python
DataScience_df.drop('employment_type', axis=1)
```

```python
DataScience_df.drop('salary', axis=1)
```

```python
#Heatmap plot
plt.figure(figsize=(12,8))
sns.heatmap(DataScience_df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
plt.figure(figsize=(5,10))
sns.countplot(x='work_year', palette='coolwarm', data=DataScience_df)
```

```python
#Countplot with Hue = Sex 
plt.figure(figsize=(12,8))
sns.countplot(x='work_year',hue='Y',palette='coolwarm', data=DataScience_df)
```
