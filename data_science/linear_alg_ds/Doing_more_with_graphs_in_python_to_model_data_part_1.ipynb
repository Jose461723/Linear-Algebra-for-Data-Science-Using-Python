```python
#Introduction
#Some of the major pros of Matplotlib are:
    #Generally easy to get started for simple plots
    #Support for custom labels and texts
    #Great control of every element in a figure
    #High-quality output in many formats
    #Very customizable in general
```

```python
#pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Basic Example
x = np.linspace(-3,3,20)
y = x**2
```

```python
x
```

```python
plt.plot(x,y,'b-')
```

```python
#Basic Matplotlib Commands 
plt.plot(x,y,'g*-')
plt.xlabel('x label')
plt.ylabel("Y label")
plt.title('Our Very First Plot')
```

```python
#Creating Multiplots on Same Canvas
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.title('Y is functionof X')

plt.subplot(1,2,2)
plt.plot(y,x,'r--')
plt.title('X is a function of Y')
```

```python
#Introduction to the Object Oriented Method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.9,0.8])
axes.plot(x,y,'b*')
```

```python
fig = plt.figure(figsize=(10,8))
axe1 = fig.add_axes([0.1,0.1,0.8,0.8])
#axe1.plot(x,y,'b-o', lw=2)
axe1.set_title('Larger Graph')

axe2 = fig.add_axes([0.1,0.1,0.2,0.3])
#axe2.plot(x,z,'r->', lw=2)
axe2.set_title('Saller Graph')
```

```python
#Subplots()
#The plt.subplots() object will act as a more automatic axis manager.
#Basic use cases:

#Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

#Now use the axes object to add stuff to plot
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
#Then you can specify the number of rows and columns when creating the subplots() object:
#Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,5))
```

```python
#Axes is an array of axes to plot on
axes
```

```python
#We can iterate through this array:
```

```python
for ax in axes:
    ax.plot(x, y, 'm')
    ax.set_xlabel('x-label')
    ax.set_ylabel('y-label')
    ax.set_title('Repeating')
    
#Display the figure object
fig
```

```python
#Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF.
#To save a figure to a file we can use the savefig method in the Figure class:
```

```python
fig.savefig("filename.png")
```

```python
#Legends, labels and titles
#Now that we have covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how decorate a figure with titles, axis labels, and legends.
```

```python
#Figures titles 
#A title can be added to each axis instance in a figure. To set the title, use the set_title method in the axes instance:
ax.set_title("title");
```

```python
#Axis labels
#Similarly, with the methods set_xlabel and set_ylabel, we can set the label of the X and Y axes:
```

```python
ax.set_xlabel("x")
ax.set_ylabel("y");
```

```python
#Legends
#You can use the label="label text" keyword argument when plots or other objects are added to the figure, and the using the legend method 
#without arguments to add the legend to the figure:
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, 'r*', label="x**2")
ax.plot(x, x**3, 'g--', label="x**3")
ax.legend(loc=4)
```

```python
#Legend options

ax.legend(loc=1) #upper right corner
ax.legend(loc=2) #upper left corner
ax.legend(loc=3) #lower left corner
ax.legend(loc=4) #lower right corner

#.. many more options are available 

#Most common to choose
ax.legend(loc=4) #let matplotlib decide the optimal location
fig
```
